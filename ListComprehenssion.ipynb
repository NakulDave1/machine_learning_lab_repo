```python
#print all the keys in lowercase if value is greater than equal to 3000

dict_car={'Maruti':100,'SUV':2000,'Audi':3000,'Baleno':4000,'Ford':6000}

l_car=[i.lower() for i in dict_car if dict_car[i]>=3000]
print(l_car)
```
